```python
import numpy as np
import pandas as pd

# 2. Define the number of samples
num_samples = 1000
print(f"Number of samples set to: {num_samples}")
import numpy as np
import pandas as pd

# 2. Define the number of samples (already defined, but including for completeness of the subtask block)
num_samples = 1000

# 3. Generate synthetic data for features
# Credit Score: Normally distributed between 300 and 850
credit_score = np.random.normal(650, 100, num_samples)
credit_score = np.clip(credit_score, 300, 850).astype(int)

# Income: Log-normally distributed to simulate varied income levels (e.g., $30,000 to $200,000)
income = np.random.lognormal(mean=10.5, sigma=0.8, size=num_samples) # Adjust mean and sigma for desired range
income = np.clip(income, 30000, 200000).astype(int)

# Loan Amount: Uniformly distributed (e.g., $5,000 to $50,000)
loan_amount = np.random.uniform(5000, 50000, num_samples).astype(int)

# Loan Term: Discrete values (12, 24, 36, 60 months)
loan_term = np.random.choice([12, 24, 36, 60], size=num_samples)

# 4. Create the target variable 'loan_approved'
# Define a logical rule for approval: higher credit_score, higher income, lower loan_amount increase probability
# Introduce some random noise to add variability

# Base probability of approval
prob_approval = (
    (credit_score / 850) * 0.4 +
    (income / 200000) * 0.3 +
    (1 - (loan_amount / 50000)) * 0.2 +
    (1 - (loan_term / 60)) * 0.1
) # Weights sum to 1.0

# Add noise and ensure probabilities are within [0, 1]
noise = np.random.normal(0, 0.1, num_samples) # Mean 0, std dev 0.1
prob_approval = np.clip(prob_approval + noise, 0, 1)

# Convert probabilities to binary outcomes (0 or 1) with a threshold
loan_approved = (prob_approval > 0.5).astype(int)

# 5. Combine all features and the target variable into a single Pandas DataFrame
df_loan = pd.DataFrame({
    'credit_score': credit_score,
    'income': income,
    'loan_amount': loan_amount,
    'loan_term': loan_term,
    'loan_approved': loan_approved
})

# 6. Display the first 5 rows of the DataFrame
print("\nFirst 5 rows of the generated DataFrame:")
print(df_loan.head())

# 7. Print descriptive statistics of the DataFrame
print("\nDescriptive statistics of the DataFrame:")
print(df_loan.describe())

from sklearn.model_selection import train_test_split

# 1. Separate features (X_loan) and target variable (y_loan)
X_loan = df_loan.drop('loan_approved', axis=1) # Features are all columns except 'loan_approved'
y_loan = df_loan['loan_approved'] # Target variable

# 3. Split X_loan and y_loan into training and testing sets
X_train_loan, X_test_loan, y_train_loan, y_test_loan = train_test_split(X_loan, y_loan, test_size=0.2, random_state=42)

# 4. Print the shapes of the resulting datasets
print(f"Shape of X_train_loan: {X_train_loan.shape}")
print(f"Shape of X_test_loan: {X_test_loan.shape}")
print(f"Shape of y_train_loan: {y_train_loan.shape}")
print(f"Shape of y_test_loan: {y_test_loan.shape}")
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
loan_model = LogisticRegression(random_state=42, solver='liblinear')

# Fit the model to the training data
loan_model.fit(X_train_loan, y_train_loan)

print("Logistic Regression model trained successfully for loan approval.")
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# 1. Use the trained loan_model to make predictions on the X_test_loan dataset
y_pred_loan = loan_model.predict(X_test_loan)

# 2 & 3. Calculate evaluation metrics
accuracy = accuracy_score(y_test_loan, y_pred_loan)
precision = precision_score(y_test_loan, y_pred_loan)
recall = recall_score(y_test_loan, y_pred_loan)
f1 = f1_score(y_test_loan, y_pred_loan)

# 4. Calculate the confusion matrix
conf_matrix = confusion_matrix(y_test_loan, y_pred_loan)

# 5. Print all calculated metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)
import matplotlib.pyplot as plt
import seaborn as sns

# Create a heatmap of the confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Rejected', 'Approved'], yticklabels=['Rejected', 'Approved'])

# Label the axes
plt.xlabel('Predicted')
plt.ylabel('Actual')

# Add a title
plt.title('Confusion Matrix Heatmap for Loan Approval Prediction')

# Display the plot
plt.show()

```
